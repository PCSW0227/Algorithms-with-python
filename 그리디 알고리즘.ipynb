```python
# chapter 03 그리디 : 당장 좋은 것만 선택하는 그리디
# 사전에 외우고 있지 않아도 풀 수 있는 묹 유형, 창의력, 즉, 아이디어가 필요한 문제다. 다시 말해, 단순히 현재 상황에서 가장 좋아 보이는 것만을 선택해도 문제를 풀 수 있는지를 파악할 수 있어야 한다.
# 거스름돈, 가게의 점원은 500원, 100원, 50원, 10원이 있다. 고객에게 1,260원을 줄 때 500원 2개, 100원 2개, 50원 1개, 10원 1개를 준다. 이 아이디어는 '가장 큰 순서대로'준 것이다.
# 즉, 그리디 알고리즘은 '가장 큰 순서대로' 혹은 '가장 작은 순서대로'와 같은 기준을 알게 모르게 제시해주기 때문에 이 아이디어를 활용하면 된다.



# 문제  : 당신은 음식점의 계산을 도와주는 점원이다. 카운터에는 거스름돈을 사용할 500원, 100원, 50원, 10원짜리 동전히 무한히 존재한다.
#         손님에게 거슬러 줘야 할 돈이 N원일 때, 거슬러 줘야 할 동전의 최소 개수를 구해라. 단, 거슬러 줄 돈 N은 항상 10의 배수다.



# 문제 정의 : 손님에게 N개의 거스름돈을 거슬러 줄 때, 500,100,50,10원을 사용하여 손님에게 거슬러 줘야 할 동전의 최소 개수를 구하자

n = 1260 #변수 설정 : 거슬러 줘야 할 돈(N원)
count = 0 #변수 설정 :사용할 동전의 개수를 카운트
coin_types = [500,100,50,10] #변수 설정 : 사용 가능한 동전의 종류를 내림차순으로 배열

for coin in coin_types: #알고리즘 실행 : coint_types의 리스트를 하나씩 순서대로 가져옵니다.
  count += n // coin #알고리즘 실행 : 동전의 개수를 최소화 하는 것은 즉, 동전의 가치가 큰 값을 먼저 사용하는 것이다. 500*2개=1,000원
  n %= coin #알고리즘 실행: 동전의 나머지를 반환(260원)

print(count) #결과 출력


# 그리디 알고리즘의 정당성 : 그리디 알고리즘으로 문제의 해법을 찾았을 때는 그 해법이 정당한지 검토해야 한다.
# 거스름돈 문제를 그리디 알고리즘으로 해결할 수 있는 이유는 가지고 있는 동전 중에서 큰 단위가 항상 작은 단위의 배수이므로 작은 단위의 동전들을 종합해 다른 해가 나올 수 없기 때문이다.
# 그리디 알고리즘을 사용할 때는 그 해법이 항상 최적의 해를 제공하는지 검증하는 것이 중요합니다. 그리디 알고리즘은 '매 순간 최적이라고 생각되는 선택을 계속해서 결정하는 방식'으로, 특정 조건에서는 최적의 해를 제공하지만 모든 경우에 그렇지는 않습니다.
#거스름돈 문제와 같은 경우, 그리디 알고리즘의 정당성은 사용되는 동전의 단위가 서로 배수 관계에 있기 때문에 성립합니다. 이 경우에는 큰 단위의 동전이 항상 작은 단위의 동전들의 조합보다 더 큰 가치를 갖기 때문에, 가장 큰 단위의 동전부터 최대한 사용하는 것이 최적의 해를 보장합니다.
#하지만 예로 든 800원을 거슬러 줘야 하는 경우처럼(동전 단위가 500원, 400원, 100원인 경우), 동전의 단위가 서로 배수 관계에 있지 않으면 그리디 알고리즘을 사용했을 때 최적의 해를 찾지 못할 수 있습니다. 이 경우에는 500원, 400원, 100원 단위의 동전이 있고, 그리디 알고리즘은 500원 1개와 100원 3개를 선택하지만, 실제 최적의 해는 400원 2개입니다.
#따라서 그리디 알고리즘을 적용하기 전에는 항상 문제의 특성을 분석하고, 그리디 알고리즘이 최적의 해를 제공할 것인지 검토하는 것이 필수적입니다.
```
